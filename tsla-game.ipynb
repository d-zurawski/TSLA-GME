```python
import yfinance as yf
import pandas as pd
ticker = "TSLA"

tesla = yf.Ticker(ticker)

historical_data = tesla.history(period="1d", start="2023-01-01", end="2023-07-03")

historical_data.reset_index(inplace=True)
tesla_data = pd.DataFrame(historical_data)

print(tesla_data.head())
```

```python
import pandas as pd
tesla_revenue = pd.DataFrame(historical_data)
print(tesla_revenue.tail(5))
```

```python
ticker = "GME"

gme = yf.Ticker(ticker)

historical_data = gme.history(period="1d", start="2023-01-01", end="2023-07-03")

historical_data.reset_index(inplace=True)

gme_revenue = pd.DataFrame(historical_data)
print(gme_revenue.head())
```

```python
print(gme_revenue.tail(5))
```

```python
import matplotlib.pyplot as plt

plt.plot(tesla_revenue.index, tesla_revenue['Close'])

plt.title("Tesla Stock Data")
make_graph = plt.show()
make_graph()
```
